## Purchasing Analysis (Gender)

```python
# Drop duplicate SN entries to get unique players
unique_players_df = purchase_data.drop_duplicates(["SN"],keep="first")

# Create df with the totals per gender, rename column
gender_count = pd.DataFrame(unique_players_df["Gender"].value_counts())
gender_count.columns = ["Total Count"]

# Set total users, Add column for percentage
total_users = gender_count["Total Count"].sum()
gender_count["Percentage"] = [round((x/total_users)*100) for x in gender_count["Total Count"]]

gender_count
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Find total number of purchases by gender.
gender_purchases = purchase_data.groupby("Gender")["Item Name"].count()
gender_purchases
```

```python
#Find average price of items by gender.
gender_avg_price = purchase_data.groupby("Gender")["Price"].mean()
gender_avg_price.round(2)
```

```python
#Find total purchase value by gender.
gender_total_purchase = purchase_data.groupby("Gender")["Price"].sum()
gender_total_purchase
```

```python
#Find average total purchase per gender
gender_avg_total_purchase = round(gender_total_purchase/gender_count["Total Count"],2)
gender_avg_total_purchase.round(2)
```

```python
# Purchasing analysis DataFrame by gender.

gender_analysis = pd.DataFrame({"Purchase Count": gender_purchases,
                                "Average Purchase Price":gender_avg_price,
                                "Total Purchase Value":gender_total_purchase,
                                "Avg Total Purchase per Person":gender_avg_total_purchase})

#Gender analysis Formatting
gender_analysis["Average Purchase Price"] = gender_analysis["Average Purchase Price"].map("${:.2f}".format)
gender_analysis["Total Purchase Value"] = gender_analysis["Total Purchase Value"].map("${:.2f}".format)
gender_analysis["Avg Total Purchase per Person"] = gender_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
gender_analysis = gender_analysis[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]


gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create bins in which data will be held. Bins are <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39 >39.

bins = [0, 9, 14, 19, 24,29,34,39,100]
age_ranges = ['<10', '10-14', '15-19', '20-24','25-29','30-34','35-39','40+']
```

```python
# Cut purchase data and place the ages into bins
pd.cut(purchase_data ["Age"], bins, labels=age_ranges)
```

```python
#Add age range column to main dataframe
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels= age_ranges)
purchase_data.head()
```

```python
#Find players percentage by age range.
groupby_age_bin = purchase_data.groupby(["SN","Age Range"])
groupby_age_bin.head()                                                            
```

```python
#Players percentage by age range.

age_group_percentage = round(purchase_data["Age Range"].value_counts()/780,2)
age_group_percentage
```

```python
# Purchase count by age range.

age_group_count = purchase_data.groupby("Age Range")["Price"]
age_group_counts= age_group_count.count()
age_group_counts
```

```python
# Average purchase price by age range.

age_group_average = purchase_data.groupby("Age Range")["Price"].mean()
age_group_average.round(2)
```

```python
#Total purchase value by age range.

age_group_total = purchase_data.groupby("Age Range")["Price"].sum()
age_group_total
```

```python
# Convert to DataFrame.

age_range_df = pd.DataFrame({"Purchase Count":age_group_counts,
                            "Average Purchase Price":age_group_average,
                            "Total Purchase Value": age_group_total,
})

#Age range formatting
age_range_df["Average Purchase Price"]=age_range_df["Average Purchase Price"].map("${:.2f}".format)
age_range_df["Total Purchase Value"]=age_range_df["Total Purchase Value"].map("${:.2f}".format)

age_range_df
```

## Purchasing Analysis (Age)

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Overall spending analysis

players_purchase_count = purchase_data.groupby("SN").count()["Price"].rename("Purchase Count")
players_average_price = purchase_data.groupby("SN").mean()["Price"].rename("Average Purchase Price")
players_total = purchase_data.groupby("SN").sum()["Price"].rename("Total Purchase Value")

#Convert to DataFrame.

total_user_data_df = pd.DataFrame({"Purchase Count":players_purchase_count,
                                   "Average Purchase Price": players_average_price,
                                   "Total Purchase Value": players_total})
#User dataframe formatting
total_user_data_df["Average Purchase Price"]=total_user_data_df["Average Purchase Price"].map("${:.2f}".format)
total_user_data_df["Total Purchase Value"]=total_user_data_df["Total Purchase Value"].map("${:.2f}".format)
total_user_data_df.head(30)
```

```python
# Total items purchased analysis.

items_purchase_count = purchase_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
items_average_price = purchase_data.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Average Purchase Price")
items_value_total = purchase_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
#unique_users= purchase_data['SN'].unique()
# Convert to DataFrame

items_purchased_df = pd.DataFrame({"Purchase Count":items_purchase_count,
                                   "Average Purchase Price":items_average_price,
                                   "Total Purchase Value":items_value_total,
                                   #"Average Total Purchase Per Person:":items_value_total/items_purchase_count,
                                  })

#Items Purchased formatting
items_purchased_df["Average Purchase Price"]=items_purchased_df["Average Purchase Price"].map("${:.2f}".format)
items_purchased_df["Total Purchase Value"]=items_purchased_df["Total Purchase Value"].map("${:.2f}".format)
items_purchased_df.head()
```
