```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.arange(1, 13, 1)
x_axis
```

```python
# Creates an exponential series of values which we can then chart
fTemps = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
fTemps
```

```python
plt.plot(x_axis, fTemps)
plt.show
```

```python
#Convert to Celsius C = (F - 32) * 0.56
cTemps = [(x-32) * 0.56 for x in fTemps]
cTemps
```

```python
plt.plot(x_axis, cTemps)
plt.show()
```

```python

# Plot both on the same chart
plt.plot(x_axis, fTemps)
plt.plot(x_axis, cTemps)
plt.show()
```

```python
f_handle,=plt.plot(x_axis,fTemps, marker='+', color='blue', label="Fahrenheit")
c_handle,=plt.plot(x_axis,cTemps, marker='s', color='red', label="Celsius")
```

```python
plt.legend(handles=[f_handle, c_handle], loc="upper right")
```

```python
# Create labels for the X and Y axis
plt.xlabel("Months")
plt.ylabel("Degrees")
```

```python
plt.show()
```
